2\. The swap function

Write a function that swap the values of two input variables x and y (whatever the type). Try to do that also without a temporary variable

```python
# with temporary variable

def swap(a,b):
    y = 0
    y = a
    a = b
    b = y
    return a,b

x1 = float(input("enter the first variable: "))
x2 = float(input("enter the second variable: "))
swap(x1,x2)
```

```python
# without temporary variable

def swap(x):
    x.reverse()
    return x

x1 = float(input("enter the first variable: "))
x2 = float(input("enter the second variable: "))
x = [x1,x2]
swap(x)
```

3\. Computing the distance

Write a function that calculates and returns the euclidean distance between two points *u* and *v*, where *u* and *v* are both 2-tuples *(x,y)*. For example, if *u=(3,0)* and *v=(0,4)*, the function should return 5

```python
import math

def distance(u,v):
    d = math.sqrt((v[1]-u[1])**2 + (v[0]-u[0])**2)
    return d

u = tuple(input("enter the first point: "))
v = tuple(input("enter the second point: "))
x1 = list(u)
x2 = list(v)
x1 = [int(i) for i in x1]
x2 = [int(j) for j in x2]
print(distance(x1,x2))
```

8\. Nested list comprehension

A Pythagorean triple is an integer solution to the Pythagorean theorem $a^2+b^2=c^2$. The first Pythagorean triple is (3,4,5). Find and put in a tuple all unique Pythagorean triples for the positive integers a, b and c less than 100.

```python
def pythagoreanTriplets(limits) : 
    c, m = 0, 2
     
    while c < limits : 
           
        for n in range(1, m) : 
            a = m * m - n * n 
            b = 2 * m * n 
            c = m * m + n * n  
            
            if c > limits : 
                break
            t = (a,b,c)
            print(t)
        m = m + 1
      
limit = 99
pythagoreanTriplets(limit)
```

9\. Normalization

Write a function that takes a tuple of numbers and returns it with the entries normalized to one

```python
a_list = list(input("enter the numbers: "))
a_list = [int(i) for i in a_list]
a_tuple = tuple(a_list)
minimum = min(a_list)
maximum = max(a_list)
d = maximum - minimum
normalized_list = list(map(lambda i: (i - minimum) / d, a_list ))
normalized_tuple = tuple(normalized_list)
print(normalized_tuple)
```
