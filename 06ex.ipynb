```python
import numpy.random as npr
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
fig = plt.figure(figsize=(8, 8))
fig, ax = plt.subplots(figsize=(9, 5))

def major_tick(x, pos):
    if (x % 1.0 == 0):
        return x
    else :
        return ""
k= np.random.normal(loc= 4, scale=1, size=100)
ax.hist(k, bins=10,histtype='stepfilled')
ax.set_xlabel('Number of random Numbers')
ax.set_ylabel('Value of the random Number')
ax.yaxis.set_major_formatter(FuncFormatter(major_tick))
ax.tick_params(which="major", axis="y", direction="inout",labelsize=10)
ax.errorbar(2,2,yerr=1)
```

```python
counts,bin_edges = np.histogram(k,10)
bin_centres = (bin_edges[:-1] + bin_edges[1:])/2.
err = np.sqrt(counts)
plt.errorbar(bin_centres, counts, yerr=err, fmt='o')

plt.show()
```

```python
from scipy.stats import norm
fig,(ax11,ax22)=plt.subplots(nrows=1,ncols=2,figsize=(14,7))
y,var,bins=ax11.hist(k,bins=10,histtype='step',label='dis')
cent=0.5*(var[1:]+var[:-1])
gauss=[]
s=1.06*k.std()*k.size**(-0.2)
for e1 in k:
    ls=np.linspace(-5,+5,200)
    rv=norm(e1,s)
    gauss.append(rv.pdf(ls))
    ax22.plot(ls,gauss[-1])
```
