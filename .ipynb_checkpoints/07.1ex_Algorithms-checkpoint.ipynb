1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rdn
import scipy.stats as stats
import pandas as pd
from scipy import linalg as la
%matplotlib inline
```

```python
x1 = rdn.normal (0, 1, 100000)
x2 = x1 + rdn.normal (0, 3, 100000)
x3 = 2*x1 + x2  
u = np.array([list(x1),list(x2), list(x3)])
cov = np.cov(u)

mu = [0,0,0]
n = 5000
X = np.random.multivariate_normal(mu, cov, n).T

l, V = la.eig(np.cov(X))
U, spectrum, Vt = np.linalg.svd(X)

l_svd = spectrum**2/(n-1)
V_svd = U

print("l EIG:", np.real_if_close(l), "\nl SVD:", l_svd, "\n")
print("V EIG:", V, "\nV SVD:", V_svd)
print('\nI risultati ottenuti con "la.eig" e "np.linalg.svd" sono gli stessi, al netto\n delle differenze strutturali fra i due metodi (es. svd ordina in modo discendente\n gli autovalori)\n')

Lambda = np.diag(np.real_if_close(l))
print (Lambda[0,0]/Lambda.trace(), ", x1: 93% della variabilità totale") 
print (Lambda[1,1]/Lambda.trace(), ", x2: intorno allo 0% della variabilità totale")
print (Lambda[2,2]/Lambda.trace(), ", x3: 7% della variabilità totale")
```

```python
Xp = np.dot(V.T, X)

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3, figsize=(13,7))
ax1.scatter(X[0,:], X[1,:], alpha=0.2)
ax2.scatter(X[0,:], X[2,:], alpha=0.2)
ax3.scatter(X[1,:], X[2,:], alpha=0.2)
ax4.scatter(Xp[0,:], Xp[1,:], alpha=0.2)
ax5.scatter(Xp[0,:], Xp[2,:], alpha=0.2)
ax6.scatter(Xp[1,:], Xp[2,:], alpha=0.2)

print("i valori in Xp[1,:] sono praticamente azzerati, come si può notare in y in ax4\n e in x in ax6, dunque la dimensione è ridotta di 1, per descrivere la variazione\n ora bastano gli assi Xp[0,:] e Xp[2,:]")
```
